```python
import time
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid_data_1000.csv', nrows=1000)
plot_df = df.groupby(['date', 'state'])['cases'].sum().reset_index() # group by date and state and sum cases

plot_df = plot_df.groupby(['state'])['cases'].sum().reset_index()

# ascending order

plot_df = plot_df.sort_values(by=['cases'], ascending=True)

# plot date aganist cases

# Create a horizontal bar plot
plot_df.plot(kind='barh', x='state', y='cases', figsize=(5, 10))

# Set plot labels and title
plt.xlabel('Total Cases')
plt.ylabel('State')
plt.title('Total COVID-19 Cases by State')
```
